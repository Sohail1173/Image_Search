```python
def add_images_to_db(images_path):
    existing_img = []

    # Initialize vector store only once
    # vector_store = Chroma(
    #     embedding_function=embed_image,  # assumed to be a callable, not the result
    #     client=client,
    #     persist_directory=persist_directory
    # )
    if not os.path.isdir('chroma_db'):
        pass

    client=PersistentClient(path="chroma_db")
    vector_store=client.get_or_create_collection("image_embeddings")

    # Check if chroma DB exists
    if os.path.exists("chroma_db"):
        print("Checking existing entries in ChromaDB")
        try:
            entities = vector_store.get(include=["metadatas"])
            print(f"Entities found: {len(entities['metadatas'])}")
            for entry in entities['metadatas']:
                if "source" in entry:
                    existing_img.append(entry["source"])
        except Exception as e:
            print(f"Error checking existing URLs: {str(e)}")

    print(f"Existing image paths in DB: {set(existing_img)}")

    # Filter new images
    new_images = [img for img in images_path if img not in set(existing_img)]

    if not new_images:
        print("No new images to add.")
        return vector_store

    # Add new images in batches
    MAX_BATCH_SIZE = 100
    for i in range(0, len(new_images), MAX_BATCH_SIZE):
        batch = new_images[i:i + MAX_BATCH_SIZE]
        try:
            vector_store.add_documents(batch)
            print(f"Vectors for batch {i} to {i + len(batch)} stored successfully...")
        except Exception as e:
            print(f"Error adding batch: {str(e)}")

    return vector_store
```
